```python
def calculate_price_pizza():
    print("Beckett Pizza Plaza 4-for-3 Offer")
    print("="*34)

    prices = []

    while len(prices)<4:
        try:
            price = float(input(f"Enter Price of Pizza #{len(prices)+1 }: "))
            if price<=0:
                print("Please enter a valid price! ")
            else:
                prices.append(price)
        except ValueError:
            print("Please enter a valid price! ")

    prices.sort()
    total_price = sum(prices)
    cheapest = min(prices)
    final_price = total_price - cheapest
    discount_percent = (cheapest/total_price)*100

    print(f"Order Total is £{final_price:.2f}, a fabulous discount of {discount_percent:.2f}%! ")

if __name__ == "__main__":
    calculate_price_pizza() 
```
