```python
import random

def guess_password():
    while True:
        password = input("Enter Your Password : ")

        if len(password) < 9:
            print("\nPassword is too short.")
            continue
    
        attempts = 3
        passed = True

        for attempts in range(0,attempts):
            num = random.randrange(1,len(password))

            Letter_check = input(f"{attempts+1} Enter letter at position {num} : ")
            if Letter_check != password[num-1]:
                print("Security Check Failed!")
                passed = False
                break

            print("Correct!")
        if passed:
            print("Security Check Passed!")
            break

if __name__ == "__main__":
    guess_password()
    
```
