```python
import sys
import os
import string

sys.argv = ['Que5_Week8.py' , 'Check.txt' , 'dictionary.txt']

def load_dictionary(dict_file):
    if not os.path.isfile(dict_file):
        print(f"Dictionary file '{dict_file}' not found.")
        sys.exit(1)
    with open(dict_file, 'r') as f:
        words = set(word.strip().lower() for word in f)
    return words

def spell_check(filename, dictionary):
    if not os.path.isfile(filename):
        print(f"File '{filename}' not found.")
        return

    with open(filename, 'r') as f:
        for line in f:
            line = line.translate(str.maketrans('', '', string.punctuation))
            for word in line.split():
                if word.lower() not in dictionary:
                    print(word)

if __name__ == "__main__":
    if len(sys.argv) != 3:
        print("Please provide the dictionary file and a text file.")
    else:
        dict_file = sys.argv[1]
        text_file = sys.argv[2]

        dictionary = load_dictionary(dict_file)
        spell_check(text_file, dictionary)
```
