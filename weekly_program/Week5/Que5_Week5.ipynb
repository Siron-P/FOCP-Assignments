```python
import sys

sys.argv = ['Que5_Week5.py' , '12.4' , '13.5' , '15' ]

args = sys.argv[1:]

if not args:
    print("No temperature values provided.")
    sys.exit(1)

temps = [float(i) for i in args]

minimum = min(temps)
maximum = max(temps)
mean = sum(temps) / len(temps)

print(f"Minimum temperature: {minimum}")
print(f"Maximum temperature: {maximum}")
print(f"Mean temperature: {mean}")
```
