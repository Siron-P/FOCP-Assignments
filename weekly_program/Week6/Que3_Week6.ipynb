```python
def prime_num(num):
    if num <= 1:
        return False
    
    for i in range(2,num):
        if num%i==0:
            return False
    return True

n = int(input("Enter any number : "))
if prime_num(n):
    print(f"{n} is prime number.")
else:
    print(f"{n} is not prime number.")
```
