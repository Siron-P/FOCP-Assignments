```python
def freq_analysis(msg):
    freq = {}
    msg = msg.lower()

    for ch in msg:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1

    result = sorted(freq.items(),key = lambda x: x[1], reverse=True)

    return result[:6]

text = input("Enter any string: ")
text = text.replace(" ","")
result = freq_analysis(text)

print("\nSIX COMMON LETTERS AND THEIR COUNT : \n")
for letter, count in result:
    print(f"{letter} : {count}")
```
