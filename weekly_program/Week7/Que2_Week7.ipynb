```python
#Letters that appear in at least one of the two words.
def occur(a,b):
    c = a|b
    c = list(c)
    return sorted(c)

#Letters that appear in both words.
def occur_twice(a,b):
    c = a&b
    c = list(c)
    return sorted(c)
#Letters that appear in either word, but not in both.
def occur_once(a,b):
    c = a^b
    c = list(c)
    return sorted(c)

a = set(input("Enter first word : "))
b = set(input("Enter second word : "))
print(f"Letters that appear in at least one of the two words are : {occur(a,b)}")
print(f"Letters that appear in both words are : {occur_twice(a,b)}")
print(f"Letters that appear in either word, but not in both are : {occur_once(a,b)}")
```
