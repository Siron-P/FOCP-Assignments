```python
country_capital = {}

print("\n --- COUNTRY AND ITS CAPITALS !! --- ")
print("Please enter 'exit' for exiting program .....\n")

while True:
    country = input("Enter the name of the country : ")

    if country.lower() == "exit":
        break

    country_title = country.title()

    if country_title in country_capital:
        print(f"The capital of {country_title} is {country_capital[country_title]}.")

    else:
        print(f"I don't know the capital of country : {country_title}.")
        capital = input("Please, enter the capital : ")
        country_capital[country_title] = capital.title()
        print(f"The capital of {country_title} is {capital.title()}")
```
