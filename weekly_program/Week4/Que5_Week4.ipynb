```python
def celsius_to_f(c):
    f = (c*1.8)+32
    return f

def fahrenheit_to_c(f):
    c = (f-32)*5/9
    return c

celsius = float(input("Enter the celsius degree : "))
fahrenheit = float(input("Enter the fahrenheit degree : "))

print(f"{celsius}°C = {celsius_to_f(celsius)}°F")
print(f"{fahrenheit}°F = {fahrenheit_to_c(fahrenheit)}°C")
```
