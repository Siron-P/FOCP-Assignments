```python
def count_letter(string):
    count_lower = 0
    count_upper = 0 
    for i in string:
        if i.isupper():
            count_upper += 1
        
        elif i.islower():
            count_lower +=1
    
    print(f"The number of uppercase letter is {count_upper}.")
    print(f"The number of lowercase letter is {count_lower}.")
    
abc = input("Enter any string: ")
count_letter(abc)
```
