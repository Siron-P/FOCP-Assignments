```python
def find():
    print("\nEnter the temperatures in celsius : \n")
    listt = []
    while True:
        temp = input(f"Temperature : ")
        if temp == "":
            break
        else:
            listt.append(float(temp))
            continue

    maximum = max(listt)
    print(f"\n Maximum value in the given temperatures = {maximum}")

    minimum = min(listt)
    print(f"\n Minimum value in the given temperatures = {minimum}")

    mean = sum(listt)/len(listt)
    print(f"\nMean of the given temperatures = {mean}\n")

find()
    
```
