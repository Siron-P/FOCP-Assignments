```python
def celsius_to_f(c):
    f = (c*1.8)+32
    return f

def fahrenheit_to_c(f):
    c = (f-32)*5/9
    return c

celsius = input("Enter the celsius degree (with 'C' at last): ")
fahrenheit = input("Enter the fahrenheit degree (with 'F' at last): ")
celsius = float(celsius[:-1])
fahrenheit = float(fahrenheit[:-1])

print(f"{celsius}°C = {celsius_to_f(celsius)}°F")
print(f"{fahrenheit}°F = {fahrenheit_to_c(fahrenheit)}°C")
```
