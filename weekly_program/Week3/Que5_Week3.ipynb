```python
BAD_PASSWORDS = ['password', 'letmein1', 'sesame123', 'hello4444', 'justinbieber']

while True:
    np = input("Enter new password: ")
    if len(np)<8 or len(np)>12 :
        print("\n------Password must be 8 to 12 characters long!------\n\n")
        continue

    p = input("Re-enter password for confirmation: ")
    if len(p)<8 or len(p)>12:
        print("\n------Password must be 8 to 12 characters long!------\n\n")
        continue

    if np!=p:
        print("\n------Password doesnot match. Please retry!------\n\n")
        continue

    if p in BAD_PASSWORDS:
        print("Please input some other password!")
        continue

    print("\n|----PASSWORD SET----|\n\n")
    break
```
