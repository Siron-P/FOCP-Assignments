```python
np = input("Enter new password: ")
count1 = len(np)
if count1<8 or count1>12:
    print("Password must be 8 to 12 characters long")
    np = input("Again,Enter new password: ")

p = input("Re-enter password for confirmation: ")
count2 = len(p)
if count2<8 or count2>12:
    print("Password must be 8 to 12 characters long")
    p = input("Again,Re-enter password for confirmation: ")

if np==p:
    print("Password set!")

else :
    print("Password does not match! Please retry.")
```
